```python
#Check for motorway signs using the actual probe data structure
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, LineString
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
%matplotlib inline
# Load Data
probe_csv_path = "/Users/kush/Chicago_Hackathon_Expanded_Datasets/23612035/23612035_probe_data.csv"
geojson_combined_path = "/Users/kush/Chicago_Hackathon_Expanded_Datasets/23612035/23612035_combined.geojson"
geojson_validation_path = "/Users/kush/Chicago_Hackathon_Expanded_Datasets/23612035/23612035_validations.geojson"
probe_df = pd.read_csv(probe_csv_path)
combined_gdf = gpd.read_file(geojson_combined_path)
validation_gdf = gpd.read_file(geojson_validation_path)
# Create Probe GeoDataFrame
probe_df["geometry"] = probe_df.apply(lambda row: Point(row["longitude"], row["latitude"]), axis=1)
probe_gdf = gpd.GeoDataFrame(probe_df, geometry="geometry", crs="EPSG:4326")
# Convert timestamp to datetime for sorting
probe_gdf['datetime'] = pd.to_datetime(probe_gdf['sampledate'])
# Extract Validation Point and Buffer
validation_point = validation_gdf.geometry.iloc[0]
buffer = validation_point.buffer(0.0002)  # ~20m buffer in degrees
# Check for motorway signs in the database near the validation point
motorway_signs_db = combined_gdf[
    (combined_gdf.geometry.type == "Point") &
    (combined_gdf["signType"].notna()) &
    (combined_gdf["signType"].str.contains("motorway", case=False, na=False)) &
    (combined_gdf.geometry.within(buffer))
]
print("=== DATABASE CHECK ===")
if len(motorway_signs_db) == 0:
    print(":rotating_light: DATABASE: No motorway signs in database near validation point")
    # This is already Scenario 1
    scenario_1 = True
else:
    print(f":white_check_mark: DATABASE: Found {len(motorway_signs_db)} motorway signs in database")
    scenario_1 = False
    # List the sign types
    for idx, sign in motorway_signs_db.iterrows():
        print(f"- Sign {idx}: {sign['signType']}")
# Get probe data near the validation point and on nearby roads
probes_near_validation = probe_gdf[probe_gdf.geometry.within(buffer)]
print(f"\nFound {len(probes_near_validation)} probe points near validation point")
# Extract roads near validation point
roads_near_validation = combined_gdf[
    (combined_gdf.geometry.type == "LineString") &
    (combined_gdf.geometry.intersects(buffer))
]
# Get probes on the road
if len(roads_near_validation) > 0:
    # Combine all road geometries
    road_union = roads_near_validation.unary_union
    # Find probes that are on or very close to the road (within 5 meters)
    probes_on_road = probe_gdf[probe_gdf.geometry.distance(road_union) < 0.00005]  # ~5m
    print(f"Found {len(probes_on_road)} probe points on nearby roads")
else:
    probes_on_road = gpd.GeoDataFrame(columns=probe_gdf.columns, geometry='geometry', crs=probe_gdf.crs)
    print("No roads found near validation point")
# Reality check based on vehicle behavior
print("\n=== REALITY CHECK BASED ON VEHICLE BEHAVIOR ===")
# If we have signs in database, check for behavior changes that would indicate sign presence
if not scenario_1:
    # For each sign, analyze vehicle behavior around it
    sign_exist_evidence = []
    for idx, sign in motorway_signs_db.iterrows():
        print(f"\nAnalyzing behavior near sign {idx} ({sign['signType']}):")
        # Create buffers before and after the sign based on heading
        # We assume vehicles approach from opposite direction of the sign heading
        sign_point = sign.geometry
        # Get all probe points near this sign (30 meters)
        sign_buffer_30m = sign_point.buffer(0.0003)
        probes_near_sign = probe_gdf[probe_gdf.geometry.within(sign_buffer_30m)]
        if len(probes_near_sign) < 3:
            print(f"  Insufficient probe data near sign {idx} to analyze behavior")
            sign_exist_evidence.append(False)
            continue
        # Group probes by traceid (vehicle)
        trace_groups = probes_near_sign.groupby('traceid')
        vehicles_analyzed = 0
        vehicles_with_behavior_change = 0
        for trace_id, trace_group in trace_groups:
            if len(trace_group) < 3:  # Need at least 3 points to analyze
                continue
            # Sort by timestamp
            trace_group = trace_group.sort_values('datetime')
            # Calculate distance to sign for each point
            trace_group['dist_to_sign'] = trace_group.geometry.apply(lambda p: p.distance(sign_point))
            # Find the closest point to the sign
            min_dist_idx = trace_group['dist_to_sign'].idxmin()
            min_dist_pos = trace_group.index.get_loc(min_dist_idx)
            # Need points before and after the closest point
            if min_dist_pos == 0 or min_dist_pos == len(trace_group) - 1:
                continue
            vehicles_analyzed += 1
            # Get speeds before and after closest approach to sign
            speeds_before = trace_group.iloc[:min_dist_pos]['speed'].values
            speeds_after = trace_group.iloc[min_dist_pos:]['speed'].values
            # Calculate average speeds
            avg_speed_before = np.mean(speeds_before)
            avg_speed_after = np.mean(speeds_after)
            speed_change = avg_speed_after - avg_speed_before
            # Get headings before and after closest approach
            headings_before = trace_group.iloc[:min_dist_pos]['heading'].values
            headings_after = trace_group.iloc[min_dist_pos:]['heading'].values
            # Calculate average headings
            avg_heading_before = np.mean(headings_before)
            avg_heading_after = np.mean(headings_after)
            heading_change = min((avg_heading_after - avg_heading_before) % 360,
                                 (avg_heading_before - avg_heading_after) % 360)
            # Check for significant changes
            significant_speed_change = abs(speed_change) > 5  # 5 km/h threshold
            significant_heading_change = heading_change > 10  # 10 degrees threshold
            if significant_speed_change or significant_heading_change:
                vehicles_with_behavior_change += 1
        # Determine if there's evidence the sign exists
        if vehicles_analyzed > 0:
            behavior_change_ratio = vehicles_with_behavior_change / vehicles_analyzed
            print(f"  Analyzed {vehicles_analyzed} vehicles passing near sign")
            print(f"  {vehicles_with_behavior_change} vehicles showed behavior changes consistent with sign presence")
            print(f"  Behavior change ratio: {behavior_change_ratio:.2f}")
            if behavior_change_ratio > 0.3:  # 30% threshold - adjust as needed
                print("  :white_check_mark: Evidence suggests sign EXISTS in reality")
                sign_exist_evidence.append(True)
            else:
                print("  :warning: Limited evidence of sign existence in reality")
                sign_exist_evidence.append(False)
        else:
            print("  :warning: No complete vehicle tracks found near sign")
            sign_exist_evidence.append(False)
    # Final determination based on behavior analysis
    if any(sign_exist_evidence):
        print("\n:white_check_mark: REALITY CHECK: Evidence suggests at least one motorway sign EXISTS in reality")
        scenario_1 = False
    elif len(sign_exist_evidence) > 0 and not any(sign_exist_evidence):
        print("\n:rotating_light: REALITY CHECK: No evidence of motorway signs existing in reality despite database entries")
        # This is a variant of Scenario 1 - signs in DB but not in reality
        scenario_1 = True
    else:
        print("\n:warning: REALITY CHECK: Insufficient data to verify sign existence")
        scenario_1 = False  # Default to database if we can't verify reality
else:
    # Already determined there are no signs in database
    pass
# Final Scenario Determination
print("\n=== FINAL DETERMINATION ===")
if scenario_1:
    print(":rotating_light: SCENARIO 1 CONFIRMED: No motorway signs exist in reality near the validation point")
    if len(motorway_signs_db) > 0:
        print("  (Note: Signs exist in database but behavior evidence suggests they don't exist in reality)")
else:
    print(":white_check_mark: NOT SCENARIO 1: Motorway signs exist in reality near the validation point")
# Visualization
fig, ax = plt.subplots(figsize=(12, 10))
# Plot validation buffer
buffer_gdf = gpd.GeoDataFrame(geometry=[buffer], crs="EPSG:4326")
buffer_gdf.plot(ax=ax, color='blue', alpha=0.3, label='Validation Buffer (20m)')
# Plot validation point
validation_gdf.plot(ax=ax, color='red', markersize=80, label='Validation Point')
# Plot roads
if len(roads_near_validation) > 0:
    roads_near_validation.plot(ax=ax, color='gray', linewidth=1.5, label='Roads')
# Plot signs
if len(motorway_signs_db) > 0:
    motorway_signs_db.plot(ax=ax, color='purple', markersize=80, label='Motorway Signs (DB)')
# Plot probe points
probes_near_validation.plot(ax=ax, color='orange', markersize=20, alpha=0.5, label='Probe Points')
# Add trace lines for vehicles (connect points by traceid)
for trace_id, group in probes_near_validation.groupby('traceid'):
    if len(group) > 1:
        # Sort by timestamp
        group = group.sort_values('datetime')
        # Get point coordinates
        x = group.geometry.x.values
        y = group.geometry.y.values
        # Plot line
        ax.plot(x, y, 'g-', linewidth=0.8, alpha=0.7)
plt.title('Motorway Sign Existence Analysis')
plt.legend()
plt.show()
```
